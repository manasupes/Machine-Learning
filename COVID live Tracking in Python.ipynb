```python
import pandas as pd

df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.head()
```

```python
df.tail()
```

```python
df = df.iloc[:-1, 1:]
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

```python
max(df['positive'])
```

Total Positive cases Based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='r', edgecolor='r')
plt.barh(total_states, df['positive'], align='center', alpha=0.5,
            color=(1,0,0),
            edgecolor=(0.5, 0.2, 0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['positive'])+100)
plt.xlabel('Positive number of cases')
plt.title('COVID cases')
plt.show()
```

###Active Number of Cases

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='r', edgecolor='k')
plt.barh(total_states, df['active'], align='center', alpha=0.5,
            color=(0.4,0.2,0.7),
            edgecolor=(0.5, 0.2, 0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['active'])+100)
plt.xlabel('Number of active cases')
plt.title('COVID cases')
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='r', edgecolor='k')
plt.barh(total_states, df['death'], align='center', alpha=0.5,
            color=(0.1,0.3,1),
            edgecolor=(0.7, 0.2, 0.8))

plt.yticks(total_states, df['state_name'])
plt.xlim(1, max(df['death'])+100)
plt.xlabel('Number of Deaths')
plt.title('COVID cases')
plt.show()
```
